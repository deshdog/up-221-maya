```python
# Import Pandas and MatPlot libraries.
import pandas as pd
import matplotlib.pyplot as plt
# Read in census data, show shape and first five rows.
income_data = pd.read_csv('census-2021-income-data.csv')
print(income_data.shape)
income_data.head()
```

```python
# Let's look at all columns in the data.
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Display a random row in the dataset.
income_data.sample()
```

```python
# Now examine the data types.
income_data.info()
```

```python
# Get rid of null columns.
income_data.columns[income_data.isna().all()].tolist()
income_data = income_data.dropna(axis=1,how="all")
# Reinspect a sample of the data.
income_data.info()
income_data.sample()
```

```python
# Identify columns to keep.
columns_to_keep = ['GEO_ID',
                   'S1901_C01_001E',
                   'S1901_C01_002E',
                   'S1901_C01_003E',
                   'S1901_C01_004E',
                   'S1901_C01_005E',
                   'S1901_C01_006E',
                   'S1901_C01_007E',
                   'S1901_C01_008E',
                   'S1901_C01_009E',
                   'S1901_C01_010E',
                   'S1901_C01_011E', 
                   'S1901_C01_012E', 
                   'S1901_C01_013E'
                  ]
# Create new dataset with all necessary columns.
income_data_final = income_data[columns_to_keep]
income_data_final.head()
```

```python
# Rename columns.
columns = list(income_data_final) # this is the same as df.columns.to_list()
income_data_final.columns = ['geo_id',
'total_households',
'percent_households_less_than_$10,000',
'percent_households_between_$10,000_and_$14,999',
'percent_households_between_$15,000_and_$24,999',
'percent_households_between_$25,000_and_$34,999',
'percent_households_between_$35,000_and_$49,999',
'percent_households_between_$50,000_and_$74,999',
'percent_households_between_$75,000_and_$99,999',
'percent_households_between_$100,000_and_$149,999', 
'percent_households_between_$150,000_and_$199,999', 
'percent_households_$200,00_and_above', 
'median_income', 
'mean_income']
# Delete the first row of the dataset.
income_data_final.drop(index = income_data_final.index[0], axis=0, inplace=True)
# See new columns applide to dataset.
income_data_final.sample(5)
```

```python
# Quick high-level analysis on data.
income_data_final = income_data_final.astype({"total_households": int})
print("Average number of households in LA County Census Tracts: ", income_data_final['total_households'].mean())
print("Median number of households in LA County Census Tracts: ", income_data_final['total_households'].median())
income_data_final['total_households'].describe()
```

```python
# See histogram of number of households in LA County census tracts.
income_data_final['total_households'].plot.hist(bins=50)
```

```python
# Sort number of households in census tracts by most to least.
total_households_data_sorted = income_data_final.sort_values(by='total_households',ascending = False)

# Make a pretty bar chart like we did in lab.
total_households_data_sorted.head(10).plot.barh(x='geo_id',
                            y='total_households', 
                            title='Top 10 Census Tracts with Highest Number of Households in Los Angeles County in 2021',
                            color='red')
```

```python
# Import geopandas library for mapping
import geopandas as gpd
# read in a geojson file downloaded from the LA Times
tracts = gpd.read_file('Census_Tracts_2020.geojson')
tracts.head()
```

```python
# Exclude all columns from dataset except FIPS and geometry.
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
# Create a FIPS Column in census tract data
tracts['FIPS'] ='06' + '037' + tracts['CT20']
tracts.head()
```

```python
# Need to create a FIPS column in the other datasest too.
income_data_final['FIPS'] = income_data_final['geo_id']
income_data_final['FIPS'] = income_data_final['geo_id'].str[9:]
income_data_final.head()
```

```python
# Join the census tract data and income data based on FIPS
tracts_income = tracts.merge(income_data_final,on="FIPS")
tracts_income.head()
```

```python
print("Density of households in LA County Census Tracts")
tracts_income.plot(figsize=(12,10),
                 column='total_households',
                 legend=True, 
                 scheme='NaturalBreaks')

# Realizing now that I can't map income breakdowns because they're in percentage form. 
# Will need to fix for analysis passed this assignment.
```
